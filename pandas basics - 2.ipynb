# work with date and time in pandas

```python
import pandas as pd
ufo = pd.read_csv("http://bit.ly/uforeports")
ufo.head()
```

```python
ufo.Time.str.slice(-5,-3).astype(int).head()    # its too old method
```

```python
ufo["Time"] = pd.to_datetime(ufo.Time)
ufo.head()
```

```python
ufo.Time.dt.hour
```

```python
ufo.Time.dt.dayofweek.head()
```

```python
ufo.Time.dt.month.head()
```

```python
ts = pd.to_datetime("10/3/1999")
```

```python
ufo.loc[ufo.Time >= ts,]
```

```python
ufo.Time.max()-ufo.Time.min()                       # used to find the difference between two mentioned dates
```

```python
ufo["year"]=ufo.Time.dt.year
ufo.head()
```

```python
ufo.year.value_counts().sort_index().plot()
```

# finding duplicate rows in pandas

```python
users.shape
```

```python
users.zip_code.duplicated().sum()
```

```python
users.duplicated().sum()
```
